```python
#Classical Part
sat_problem2 = [[-1,-2,3]]

def classical_sat_checker(sat_formula,quantum_solution):
    for sat_clause in sat_formula:
        for k in range(len(quantum_solution)):
            result = []
            for variable,quant_variable in zip(sat_clause,list(quantum_solution[k])):
                if variable>0 and int(quant_variable)>0:
                    result.append('yes')
                elif variable<0 and int(quant_variable)==0:
                    result.append('yes')
                else:
                    result = result
            if 'yes' in result:
                sat_solution = quantum_solution[k]
    return(sat_solution)

classical_soln = classical_sat_checker(sat_problem2,quant_soln)
print(classical_soln)                
            
            
                
```
